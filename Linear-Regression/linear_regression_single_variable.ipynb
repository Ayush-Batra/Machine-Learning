```python
#importing important libraries
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
```

```python
#reading csv file
data = pd.read_excel("/home/ayush/Downloads/slr08.xls")
data.columns
x = data['X']
y = data['Y']
data.head()
```

```python
#plotting the data
plt.xlabel("Experience in years")
plt.ylabel("Salary monthly")
plt.title("Experience VS Salary")
plt.scatter(x,y)
```

```python
#splitting the data into training and testing set
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
#plotting the training set
plt.xlabel("Experience in years")
plt.ylabel("Salary monthly")
plt.title("Experience VS Salary")
plt.scatter(x_train,y_train)
```

```python
#plotting the testing set
plt.xlabel("Experience in years")
plt.ylabel("Salary monthly")
plt.title("Experience VS Salary")
plt.scatter(x_test,y_test)
```

```python
#In this block the calculation takes place to calculate theta_1 and theta_0
theta_0=0
theta_1=0
m = x_train.size
learning_rate = 0.005
iteration = 10000
#gradeint descent for convergence
for i in range(iteration):
    hypothesis = theta_0 +(theta_1*x_train) #our function 
    differ_theta_0  = (sum(hypothesis - y_train))/m  #function differentiated w.r.t theta_0
    differ_theta_1  = (sum((hypothesis - y_train)*x_train))/m #function differentiated w.r.t theta_1
    theta_0 = theta_0 - (learning_rate*differ_theta_0)  #to update theta_0
    theta_1 = theta_1 - (learning_rate*differ_theta_1)  #to update theta_0
print(theta_0, theta_1)
```

```python
#To plot the linear regressive function and training points
hypothesis = theta_0 +(theta_1*x_train)
plt.scatter(x_train,y_train)
plt.xlabel("Experience in years")
plt.ylabel("Salary monthly")
plt.title("Experience VS Salary")
plt.plot([min(x_train), max(x_train)],[min(hypothesis), max(hypothesis)]) #our linear regressive function
```

```python
#To plot the linear regressive function and testing points
hypothesis = theta_0 +(theta_1*x_test)
plt.scatter(x_test,y_test)
plt.xlabel("Experience in years")
plt.ylabel("Salary monthly")
plt.title("Experience VS Salary")
plt.plot([min(x_test), max(x_test)],[min(hypothesis), max(hypothesis)]) #our linear regressive function
```

```python
#additonal step to calculate the error 
value_calculated = theta_0 +(theta_1*x_test)
value_expected = y_test
error = abs((value_calculated-value_expected)/value_expected)
total_error = (sum(error)/error.size)
total_error
```
