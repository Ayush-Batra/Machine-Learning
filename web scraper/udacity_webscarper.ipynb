```python
from selenium import webdriver
from bs4 import BeautifulSoup
from webdriver_manager.firefox import GeckoDriverManager
import pandas as pd
import requests
import json
```

```python
driver = webdriver.Firefox(executable_path="/home/ayush/Downloads/geckodriver")
```

```python
link = "https://www.udacity.com/courses/all"
driver.get(link)
content = driver.page_source
soup = BeautifulSoup(content)
```

```python
data = pd.DataFrame(columns=["Name","Rating","Organization"])
name = []
rating = []
organization = []
link =[]
summary =[]
skills=[]
```

```python
for a in soup.findAll('li', attrs={'class':'card-list_catalogCardListItem__aUQtx'}):
    course = a.find('h2')
    name.append(course.get_text())
    rate=a.find('ul', attrs={'class':'card_stats__StvYz'})
    rating.append(rate.get_text())
    provider = a.find('h3')
    organization.append(provider.get_text())
    location = a.find('a',href = True)
    link.append(("https://www.udacity.com/"+location['href']))
    about = a.find('p')
    summary.append(about.get_text())
    skill = a.find('section')
    skills.append(skill)
    
```

```python
df = pd.DataFrame({'Course_Name':name,'Rating':rating,'Organization':organization,'Link':link,'About':summary,'Skills':skills}) 
```
